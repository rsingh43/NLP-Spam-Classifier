# Imports

```python
import os

import pandas as pd

from IPython.display import display
```

# Parameters

```python
data_dirname = "../SMS-Spam/original"
data_basename = "spam.csv.gz"
results_dirname = "../SMS-Spam"
results_basename = "data.csv.gz"
```

# Data Loading

```python
filename = os.path.join(data_dirname, data_basename)
raw_df = pd.read_csv(filename, compression="gzip", encoding="latin-1")

with pd.option_context("display.max_rows", 10, "display.max_columns", None):
    display(raw_df)
```

# Data Cleaning

```python
data_df = raw_df[["v2"]].rename(columns={"v2":"text"})
data_df["spam"] = raw_df["v1"] == "spam"

with pd.option_context("display.max_rows", 50, "display.max_columns", None, "display.max_colwidth", None):
    display(data_df)
```

# Save Data

```python
filename = os.path.join(results_dirname, results_basename)
print(filename)
data_df.to_csv(filename, index=False, compression="gzip")
```
